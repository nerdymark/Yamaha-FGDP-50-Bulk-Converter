```python
"""
Bulk File Converter for the Yamaha FGDP-50

Requirements for loading audio files:
Sample format: WAV, AIFF (44.1 kHz, 16-bit, mono/stereo)
Max. time: Up to approx. 20 seconds per file, approx. 300 seconds (stereo)/600 seconds (mono) in total

"""

import os
import subprocess
from tqdm import tqdm


# Bulk convert every file in the folders folders from mp3 to wav

usb_drive = '/Volumes/NO NAME'

folders = [
    '/Users/mark/Downloads/Unsorted/Warner Bros. Sound Effects Library (1400+ Sound FX) [RH]',
    '/Users/mark/Downloads/Unsorted/Sound Ideas Sound Effects Library 1000 Series (28 CDs)'
]

out_folder = '/Users/mark/Downloads/Unsorted/Converted'
# out_folder = usb_drive

ffmpeg_settings = {
    'sample_rate': '44100',
    'bit_rate': '16',
    'channels': '2'
}

# Make the out_folder if it doesn't exist

if not os.path.exists(out_folder):
    os.makedirs(out_folder)

folder_count = 0
for folder in folders:
    for subfolder in os.listdir(folder):
        folder_count += 1

# Convert every mp3 file in the folders to wav
for folder in folders:
    for subfolder in tqdm(os.listdir(folder), total=folder_count):
        print(f'Converting {subfolder}...')
        # Skip files, only convert folders
        # is subfolder a file?
        if os.path.isfile(os.path.join(folder, subfolder)):
            continue
        else:
            for file in os.listdir(os.path.join(folder, subfolder)):
                if file.endswith('.mp3'):
                    in_file = os.path.join(folder, subfolder, file)
                    # Put folder name into file name
                    out_file = os.path.join(out_folder, subfolder + '_' + file.replace('.mp3', '.wav'))
                    subprocess.run(['ffmpeg', '-i',
                                    in_file, '-ar',
                                    ffmpeg_settings['sample_rate'],
                                    '-ac',
                                    ffmpeg_settings['channels'],
                                    '-ab', ffmpeg_settings['bit_rate'],
                                    out_file])
                    print(f'Converted {file} to {out_file}')
```
